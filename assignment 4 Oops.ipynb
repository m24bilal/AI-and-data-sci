1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python

class BankAccount:
    def __init__(self, account_holder_name, account_number, balance=0):
        self.account_holder_name = account_holder_name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
            print(f"New Balance: {self.balance}")
        else:
            print("Deposit amount must be greater than 0.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be greater than 0.")
        elif amount > self.balance:
            print("Insufficient balance! Withdrawal not allowed.")
        else:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
            print(f"Remaining Balance: {self.balance}")

    def display_account(self):
        print("\nAccount Details")
        print("Account Holder:", self.account_holder_name)
        print("Account Number:", self.account_number)
        print("Balance:", self.balance)


# Example Usage
account = BankAccount("Muhammad bilal", "123456789", 5000)

account.display_account()

account.deposit(2000)

account.withdraw(3000)

account.withdraw(5000)  # Insufficient balance
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"

    def display_result(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)
        print("Grade:", self.calculate_grade())


# Example Usage
student1 = Student("bilal", 85)
student1.display_result()

print()

student2 = Student("muhammad", 55)
student2.display_result()
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, consumer_name, units_consumed):
        self.consumer_name = consumer_name
        self.units_consumed = units_consumed

    def calculate_bill(self):
        units = self.units_consumed

        if units <= 100:
            bill = units * 15      # Rs. 15 per unit
        elif units <= 200:
            bill = (100 * 15) + ((units - 100) * 20)
        elif units <= 300:
            bill = (100 * 15) + (100 * 20) + ((units - 200) * 25)
        else:
            bill = (100 * 15) + (100 * 20) + (100 * 25) + ((units - 300) * 30)

        return bill

    def display_bill(self):
        print("Consumer Name:", self.consumer_name)
        print("Units Consumed:", self.units_consumed)
        print("Electricity Bill: Rs.", self.calculate_bill())


consumer = ElectricityBill("bilal", 300)

consumer.display_bill()
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def calculate_total_salary(self):
        bonus = self.basic_salary * 0.10      # 10% Bonus
        tax = self.basic_salary * 0.05        # 5% Tax Deduction
        total_salary = self.basic_salary + bonus - tax
        return total_salary

    def display_salary(self):
        print("Employee Name:", self.name)
        print("Basic Salary:", self.basic_salary)
        print("Total Salary:", self.calculate_total_salary())


# Example Usage
employee = Employee("bilal", 60000)
employee.display_salary()
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod
import math


class Shape(ABC):

    @abstractmethod
    def area(self):
        pass



class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2



class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width



circle = Circle(5)
rectangle = Rectangle(10, 4)

print("Circle Area:", circle.area())
print("Rectangle Area:", rectangle.area())
```

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def determine_priority(self):
        serious_diseases = [
            "Heart Attack",
            "Stroke",
            "Cancer",
            "Kidney Failure",
            "COVID-19"
        ]

        if self.age > 60 or self.disease in serious_diseases:
            return "Emergency"
        else:
            return "Normal"

    def display_patient_info(self):
        print("Patient Name:", self.name)
        print("Age:", self.age)
        print("Disease:", self.disease)
        print("Priority:", self.determine_priority())


# E
patient1 = HospitalPatient("bilal", 65, "Fever")
patient2 = HospitalPatient("Ahmed", 35, "Heart Attack")

patient1.display_patient_info()
print()
patient2.display_patient_info()

```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item_name, price):
        self.items.append((item_name, price))
        print(f"{item_name} added. Price: Rs. {price}")

    def calculate_total(self):
        total = sum(price for item, price in self.items)
        return total

    def apply_discount(self):
        total = self.calculate_total()

        # 10% discount if total exceeds Rs. 5000
        if total > 5000:
            discount = total * 0.10
            final_bill = total - discount
        else:
            discount = 0
            final_bill = total

        return discount, final_bill

    def display_bill(self):
        print("\nShopping Cart")
        print("-" * 30)

        for item, price in self.items:
            print(f"{item}: Rs. {price}")

        total = self.calculate_total()
        discount, final_bill = self.apply_discount()

        print("-" * 30)
        print("Total Bill: Rs.", total)
        print("Discount: Rs.", discount)
        print("Final Bill: Rs.", final_bill)


# Example Usage
cart = ShoppingCart()

cart.add_item("Shoes", 3000)
cart.add_item("Shirt", 1500)
cart.add_item("Watch", 2000)

cart.display_bill()
```

9. Create a class 'MobilePhone' with private attribute battery percentage. Use getter and setter
methods to update battery with validation (0–100 only).

```python
class MobilePhone:
    def __init__(self, battery_percentage):
        self.__battery_percentage = 0
        self.set_battery_percentage(battery_percentage)

    # Setter Method
    def set_battery_percentage(self, battery_percentage):
        if 0 <= battery_percentage <= 100:
            self.__battery_percentage = battery_percentage
        else:
            print("Error: Battery percentage must be between 0 and 100.")

    # Getter Method
    def get_battery_percentage(self):
        return self.__battery_percentage


# Example Usage
phone = MobilePhone(80)

print("Battery:", phone.get_battery_percentage(), "%")

phone.set_battery_percentage(95)
print("Updated Battery:", phone.get_battery_percentage(), "%")

phone.set_battery_percentage(120)  # Invalid value
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:
    def __init__(self):
        self.books = {
            "Python Programming": True,
            "Data Structures": True,
            "Machine Learning": True,
            "Database Systems": True
        }

    # Borrow a book
    def borrow_book(self, book_name):
        if book_name in self.books:
            if self.books[book_name]:
                self.books[book_name] = False
                print(f"You have borrowed '{book_name}'.")
            else:
                print(f"Sorry, '{book_name}' is currently not available.")
        else:
            print(f"'{book_name}' does not exist in the library.")

    # Return a book
    def return_book(self, book_name):
        if book_name in self.books:
            self.books[book_name] = True
            print(f"Thank you for returning '{book_name}'.")
        else:
            print(f"'{book_name}' does not belong to this library.")

    # Display available books
    def show_books(self):
        print("\nAvailable Books:")
        for book, available in self.books.items():
            status = "Available" if available else "Issued"
            print(f"{book} - {status}")


# Example Usage
library = Library()

library.show_books()

library.borrow_book("Python Programming")
library.borrow_book("Python Programming")  # Already issued

library.return_book("Python Programming")

library.show_books()
```

```python
Write a Python program that creates a class called Bank with properties for
bank names and branches. Include methods to add a branch, remove a branch,
and display all branches. Create an instance of the Bank class and perform
operations to add and remove branches.
```

```python
class Bank:
    def __init__(self, bank_name):
        self.bank_name = bank_name
        self.branches = []

    def add_branch(self, branch_name):
        self.branches.append(branch_name)
        print(f"Branch '{branch_name}' added.")

    def remove_branch(self, branch_name):
        if branch_name in self.branches:
            self.branches.remove(branch_name)
            print(f"Branch '{branch_name}' removed.")
        else:
            print(f"Branch '{branch_name}' not found.")

    def display_branches(self):
        print(f"\nBranches of {self.bank_name}:")
        if self.branches:
            for branch in self.branches:
                print(f"- {branch}")
        else:
            print("No branches available.")


# Create an instance of Bank
bank = Bank("National Bank")

# Add branches
bank.add_branch("Karachi Branch")
bank.add_branch("Lahore Branch")
bank.add_branch("Islamabad Branch")

# Display branches
bank.display_branches()

# Remove a branch
bank.remove_branch("Lahore Branch")

# Display branches again
bank.display_branches()
```

```python
import math

# Base class
class Shape:
    def area(self):
        print("Area method not implemented.")


# Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


# Subclass: Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


# Create instance of Circle
circle = Circle(5)
print(f"Area of Circle: {circle.area():.2f}")

# Create instance of Triangle
triangle = Triangle(10, 6)
print(f"Area of Triangle: {triangle.area():.2f}")
```

```python
from abc import ABC, abstractmethod
import uuid

# -------------------- Product (Abstract Class) --------------------
class Product(ABC):
    def __init__(self, name, price, stock):
        self.name = name
        self.__price = price   # encapsulation
        self.stock = stock

    def get_price(self):
        return self.__price

    def set_price(self, new_price):
        self.__price = new_price

    @abstractmethod
    def apply_discount(self):
        pass


# -------------------- Subclasses --------------------
class AC(Product):
    def apply_discount(self):
        return self.get_price() * 0.90   # 10% discount


class Refrigerator(Product):
    def apply_discount(self):
        return self.get_price() * 0.85   # 15% discount


class WashingMachine(Product):
    def apply_discount(self):
        return self.get_price() * 0.80   # 20% discount


# -------------------- Customer Class --------------------
class Customer:
    def __init__(self, name):
        self.name = name
        self.cart = []

    def add_to_cart(self, product, quantity):
        if product.stock >= quantity:
            self.cart.append((product, quantity))
            product.stock -= quantity
            print(f"{product.name} added to cart.")
        else:
            print(f"Sorry! {product.name} is out of stock.")

    def view_cart(self):
        print("\nCart Items:")
        for product, qty in self.cart:
            print(f"{product.name} x {qty}")


# -------------------- Payment & Installment --------------------
class Payment:
    def __init__(self, total):
        self.total = total

    def full_payment(self):
        print(f"Paid Full Amount: {self.total:.2f}")

    def installment(self, months):
        monthly = self.total / months
        print(f"Installment Plan: {months} months, {monthly:.2f}/month")


# -------------------- Invoice --------------------
class Invoice:
    TAX = 0.18

    def __init__(self, customer):
        self.customer = customer
        self.id = uuid.uuid4()

    def generate(self):
        print("\n------ INVOICE ------")
        print(f"Invoice ID: {self.id}")
        print(f"Customer: {self.customer.name}")

        subtotal = 0

        for product, qty in self.customer.cart:
            price = product.apply_discount()
            total_price = price * qty
            subtotal += total_price
            print(f"{product.name} x{qty} = {total_price:.2f}")

        tax = subtotal * self.TAX
        grand_total = subtotal + tax

        print(f"Subtotal: {subtotal:.2f}")
        print(f"Tax (18%): {tax:.2f}")
        print(f"Total: {grand_total:.2f}")

        # Save to file
        with open("invoice.txt", "a") as f:
            f.write(f"\nInvoice ID: {self.id}\n")
            f.write(f"Customer: {self.customer.name}\n")
            f.write(f"Total: {grand_total:.2f}\n")
            f.write("----------------------\n")

        return grand_total


# -------------------- Shop Simulation --------------------
# Create products
ac = AC("AC", 100000, 5)
fridge = Refrigerator("Refrigerator", 80000, 3)
wm = WashingMachine("Washing Machine", 50000, 2)

# Customer
customer = Customer("Bilal")

# Shopping flow
customer.add_to_cart(ac, 1)
customer.add_to_cart(fridge, 1)
customer.add_to_cart(wm, 1)

customer.view_cart()

# Generate invoice
invoice = Invoice(customer)
total_amount = invoice.generate()

# Payment
payment = Payment(total_amount)

print("\nChoose Payment Method:")
print("1. Full Payment")
print("2. Installments")

choice = input("Enter choice: ")

if choice == "1":
    payment.full_payment()
else:
    months = int(input("Enter number of months: "))
    payment.installment(months)
```
