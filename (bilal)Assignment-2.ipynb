# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
user = int(input("enter your marks"))
if user >= 90:
    print("A")
elif user >= 80:
    print("B")
elif user >= 70:
    print("C")
elif user >= 60:
    print("D")
else:
    print("Fail")
    
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day_Number = int(input("Enter A Day Number:"))
if day_Number == 1:
   print("Monday")
elif day_Number == 2:
   print("Tuesday")
elif day_Number == 3:
   print("Wednesday")
elif day_Number == 4:
   print("Thursday")
elif day_Number == 5:
   print("Friday")
elif day_Number == 6:
   print("Saturday")
elif day_Number == 7:
   print("Sunday")
else:
   print("invalid number")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter a month number:"))
if month == 12 or month == 1 or month == 2:
   print ("winter")
elif month == 3 or month == 4 or month == 5:
   print ("spring")
elif month == 6 or month == 7 or month == 8:
   print ("summer")
else:
   print("Autumn")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
user = (input("enter operator:"))
if user == "+":
    print ("Addition")
elif user == "-":
    print("substraction")
elif user == "*":
    print("multiplication")
elif user == "/":
    print("division")
else:
    print("invalid operator")
```
